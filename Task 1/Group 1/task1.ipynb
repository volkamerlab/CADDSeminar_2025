```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import rdkit
import mordred
```

```python
from mordred import descriptors
calc = mordred.Calculator(descriptors, ignore_3D=True)
len(calc)
```

```python
df = pd.read_csv(".data/standardized_AqSolDB.csv")
```

```python
df.head()
```

```python
df.drop(columns="Unnamed: 0", inplace=True)
df.head()
```

```python
df_reordered = df[['SMILES', 'Standardized_SMILES', 'LogS']]
df_reordered.head()
```

```python
mols = [rdkit.Chem.MolFromSmiles(smi) for smi in df["Standardized_SMILES"]]
mols = calc.pandas(mols=mols)
```

```python
mols.head()
```

```python
df_descriptors = pd.concat([df_reordered, mols], axis=1)
df_descriptors.head()
```

```python
df_descriptors.to_csv("./data/standardized_AqSolDB_descriptors.csv")
```
